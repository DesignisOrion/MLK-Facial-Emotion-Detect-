# Facial Emotion Detection

This software allows you to detect emotions via facial expressions shown in camera. The goal of the project was to learn more about the deep learning using OpenCV and DeepFace. Matplotlib was used to plot the images of Martin Luther King Jr. within the notebook. Within this project I was able to get an analsis of MLK age, gender and emotion. I also was able to compare two photos to determine an answer of one to be true, the other comparrison false. Enjoy.


```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
# analyzing the face of martin luther king, jr.

analysis = DeepFace.analyze(img_path = "mlksmile.jpg", actions = ["age", "gender", "emotion", "race"])
print(analysis)
```

```python
# verifying that the two images are the same person.
DeepFace.verify("mlksmile.jpg","mlksmile2.jpg") 
```

```python
# As you can see, the two pictures in this cell is "false" because its not the same person.
DeepFace.verify("mlksmile.jpg","ransmile.jpg")
```

```python
mlkdets = DeepFace.analyze(img_path = "mlksmile2.jpg", actions = ['age', 'gender', 'race', 'emotion'])
print(mlkdets["age"]," years old ",mlkdets["dominant_race"]," ",mlkdets["dominant_emotion"]," ", mlkdets["gender"])
```
